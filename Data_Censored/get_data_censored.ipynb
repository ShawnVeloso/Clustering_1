# Converting R to Python

```python
import os
import pandas as pd

# Get the current script's directory
current_dir = os.path.dirname(os.path.abspath(os.getcwd()))  # Use os.getcwd() if running in Jupyter

# Define the file name
file_name = "data_censored.csv"

# Create the full file path
file_path = os.path.join(current_dir, file_name)

# Load the dataset
df = pd.read_csv(file_path)

# Save the dataset again to confirm it works
output_path = os.path.join(current_dir, "csv-files", file_name)  # Save inside "csv-files" subfolder
os.makedirs(os.path.dirname(output_path), exist_ok=True)  # Create folder if it doesn't exist
df.to_csv(output_path, index=False)

# Confirm the file is saved
print(f"CSV file saved at: {output_path}")
```
